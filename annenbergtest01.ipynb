```python
import pandas as pd
import numpy as np
fileName="activities-export.csv"
df = pd.read_csv(fileName,encoding="ISO-8859-1") 
df
```

```python
df2 = df.set_index(df["ID"])
del df2["ID"] 
df2
```

```python
inputId = int(input("input program's id:"))
inputRow = df2.loc[inputId,:]
inputRow
```

```python
from fuzzywuzzy import fuzz
def getSim(row):
    #print(row)
    TitleSim = fuzz.ratio(str(inputRow.Title),str(row.Title))
    CategorySim = fuzz.ratio(str(inputRow.Category),str(row.Category))
    LevelSim = fuzz.ratio(str(inputRow.Level),str(row.Level))
    DescriptionSim = fuzz.ratio(str(inputRow.Description),str(row.Description))
    #LinkSim = fuzz.ratio(inputRow.Link,row.Link)
    #CenterSim = fuzz.ratio(inputRow.Center, row.Center)
    return CategorySim * 10 + LevelSim * 10 + TitleSim * 2 + DescriptionSim * 2
```

```python
df2["sim"] = df2.apply(getSim,axis = 1 )  # 2D
df3 = df2[ ["sim"] ]
df3
```

```python
df4 = df3.sort_values(by="sim",ascending=False,axis=0)
df4
```
